# **Preprocessing our data** 

```python
df_train['Arrival Delay in Minutes'].isna().sum()
```

```python
# handling missing values
df_train['Arrival Delay in Minutes'] = df_train['Arrival Delay in Minutes'].fillna(df_train['Arrival Delay in Minutes'].mean())
df_train['Arrival Delay in Minutes'].isna().sum()
```

```python
# dropping id
df_train= df_train.drop(['id'], axis=1)
```

```python
# correlated features using seaborn 
plt.figure(figsize=(16, 6))

map = sns.heatmap(df_train.corr(), vmin=-1, vmax=1, annot=True)

map.set_title('Correlation Heatmap', fontdict={'fontsize':12}, pad=12);
```

```python
X = df_train.drop('satisfaction',axis=1)
y = df_train[['satisfaction']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.compose import ColumnTransformer,make_column_selector
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.15,random_state=69)
```

```python
OHE=OneHotEncoder(drop='first')

standard_scaler=StandardScaler()                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        

column_transformation = ColumnTransformer([
    ('categorical_encoder',OHE,make_column_selector(dtype_include='object')),
    ('numerical_encoder',standard_scaler,make_column_selector(dtype_include='float64'))
],remainder='passthrough')
```

```python
X_train = column_transformation.fit_transform(X_train)
X_test = column_transformation.transform(X_test)
```

```python
print("-"*100)
for i in column_transformation.transformers_:
    print(i)
    if(i[0]!='remainder'):
        print(i[1].get_feature_names_out())   
    print("-"*100)
```
